```python
import numpy as np
import pandas as pd
```

```python
#load datasets
stats = pd.read_csv("datasets/nba_2017_players_stats.csv") 
salary = pd.read_csv("datasets/nba_2017_salary.csv")
#merge data sets based on name, drop na values
df1 = stats.merge(salary, on='NAME')

df2 = df1.dropna(axis=1)
df1
```

```python
#target vector
y = np.array(df2["SALARY"])
x = np.array(df2["POINTS"])
```

```python
print(np.amax(x))
```

```python
def fit_linear(x,y):
    xm = np.mean(x)
    ym = np.mean(y)
    syy = np.mean((y-ym)**2)
    syx = np.mean((y-ym)*(x-xm))
    sxx = np.mean((x-xm)**2)
    beta1 = syx/sxx
    beta0 = ym - beta1*xm
    yhat=beta0+beta1*x
    rss = np.sum((y-yhat)**2)
    tss = np.sum((y-ym)**2)
    rsq = 1-rss/tss
    
    
    return beta0, beta1, rsq
```

```python
beta0, beta1, rsq = fit_linear(x,y)
print(fit_linear(x,y))

yhat = beta0+beta1*x
```

```python
import matplotlib.pyplot as plt  
yhat = beta1*x + beta0
plt.scatter(x,y,s=8)                    
plt.plot(x,yhat,'-',linewidth=1)  
plt.xlabel('POINTS PER GAME')
plt.ylabel('Salary')
plt.title("LS Fit Solution ")
plt.show()

print("R^2 = " + str(rsq))
print("Beta0 = " + str(beta0) +"Beta1 = " + str(beta1))
```

```python
def compute_cost(x,y,beta0,beta1,m):
    #Write your code in place of None. Cost can be calculated using a single line of code
    cost=np.sum(((beta0 + x * beta1) - y) **2) / (2*m)
    
    return cost
```

```python
def gradient_descent(x,y,learning_rate,beta0,beta1,m,num_iters):
    # In place of None, write the updated value of beta0 in temp0 and of beta1 in temp1
    for i in range(num_iters):
        temp0 =   beta0 - (learning_rate/m) * np.sum((beta0 +  x * beta1)- y)
        temp1 =   beta1 - (learning_rate/m) * (np.sum(((beta0 + x * beta1)-y) * x) )
        beta0 =temp0
        beta1= temp1
            
    return beta0,beta1    
```

```python
m=x.shape[0]
beta0,beta1 = gradient_descent(x,y,0.001,0,0,m,100000)
```

```python
yhat = beta1*x + beta0
plt.scatter(x,y,s=8)                    
plt.plot(x,yhat,'-',linewidth=1)  
plt.xlabel('POINTS PER GAME')
plt.ylabel('Salary')
plt.title("Minimizing Cost")
plt.show()

ym = np.mean(y)
rss = np.sum((y-yhat)**2)
tss = np.sum((y-ym)**2)
rsq = 1-rss/tss
print("R^2 = " + str(rsq))
print("Beta0 = " + str(beta0) +"Beta1 = " + str(beta1))
```
